```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Charger les données
df = pd.read_csv('C:/Projet_ML/data/train.csv')

# Afficher les 5 premières lignes
df.head()
```

```python
# Suppression de la colonne 'id' si elle existe
if 'id' in df.columns:
    df = df.drop(columns=['id'])

# Afficher les 5 premières lignes du dataframe après suppression de la colonne 'id'
df.head()
```

```python
# Compter les valeurs manquantes
missing_values = df.isnull().sum()
print("Valeurs manquantes par colonne :\n", missing_values)

# Visualisation des valeurs manquantes
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title('Valeurs Manquantes')
plt.show()
```

```python
plt.figure(figsize=(15, 8))
sns.boxplot(data=df, orient='h', palette='Set2')
plt.title("Valeurs aberrantes (Outliers)")
plt.show()
```

```python
# Afficher la matrice de corrélation
plt.figure(figsize=(15, 10))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matrice de Corrélation')
plt.show()
```

```python
# Suppression des paires de variables fortement corrélées
correlation_matrix = df.corr().abs()
upper_triangle = pd.DataFrame(np.triu(correlation_matrix, k=1), columns=correlation_matrix.columns, index=correlation_matrix.index)

# Suppression des colonnes fortement corrélées (corrélation > 0.85)
high_correlation = [column for column in upper_triangle.columns if any(upper_triangle[column] > 0.85)]
df = df.drop(columns=high_correlation)

print(f"Colonnes supprimées pour cause de multicolinéarité: {high_correlation}")
df.head()
```

```python
# **Traitement des Outliers**
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Afficher les boxplots avant le traitement des outliers
plt.figure(figsize=(15, 8))
sns.boxplot(data=df, orient='h', palette='Set2')
plt.title("Valeurs aberrantes (Outliers) avant traitement")
plt.show()

# **Traitement des outliers (méthode IQR)**
for column in df.select_dtypes(include=[np.number]).columns:
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df[column] = np.where(df[column] < lower_bound, Q1, df[column])
    df[column] = np.where(df[column] > upper_bound, Q3, df[column])

# **Afficher les boxplots après le traitement des outliers**
plt.figure(figsize=(15, 8))
sns.boxplot(data=df, orient='h', palette='Set2')
plt.title(" Valeurs aberrantes (Outliers) après traitement")
plt.show()
```

```python
from sklearn.model_selection import train_test_split

# Séparation des données en X et y
X = df.drop(columns='smoking')
y = df['smoking']

# Division train/test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

```python
from imblearn.over_sampling import SMOTE

# Application de SMOTE
smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier

# Création des pipelines
pipelines = {
    'Logistic Regression': Pipeline([
        ('scaler', StandardScaler()),
        ('select_features', SelectKBest(score_func=f_classif, k=10)),
        ('classifier', LogisticRegression(max_iter=500, random_state=42))
    ]),
    'Random Forest': Pipeline([
        ('scaler', StandardScaler()),
        ('select_features', SelectKBest(score_func=f_classif, k=10)),
        ('classifier', RandomForestClassifier(random_state=42))
    ]),
    'Decision Tree': Pipeline([
        ('scaler', StandardScaler()),
        ('select_features', SelectKBest(score_func=f_classif, k=10)),
        ('classifier', DecisionTreeClassifier(random_state=42))
    ])
}
```

```python
from sklearn.model_selection import GridSearchCV

# Paramètres optimisés pour chaque modèle
param_grids = {
    'Logistic Regression': {
        'classifier__C': [0.01, 0.1, 1, 10, 100],
        'classifier__penalty': ['l2'],
        'classifier__solver': ['lbfgs', 'liblinear']
    },
    'Random Forest': {
        'classifier__n_estimators': [50, 100, 150],
        'classifier__max_depth': [10, 20, None]
    },
    'Decision Tree': {
        'classifier__max_depth': [5, 10, 20, None],
        'classifier__min_samples_split': [2, 5, 10]
    }
}

# Dictionnaire pour stocker les meilleurs modèles
best_models = {}

# Boucle sur chaque modèle et application de GridSearchCV
for model_name, pipeline in pipelines.items():
    print(f"Optimisation des hyperparamètres pour le modèle : {model_name}")
    
    # Récupérer les paramètres spécifiques au modèle
    param_grid = param_grids.get(model_name, {})
    
    # Configurer GridSearchCV
    grid_search = GridSearchCV(pipeline, param_grid, cv=5, scoring='accuracy', n_jobs=-1)
    
    # Entraînement du modèle avec recherche d'hyperparamètres
    grid_search.fit(X_train_resampled, y_train_resampled)
    
    # Sauvegarde du meilleur modèle
    best_models[model_name] = grid_search.best_estimator_
    
    print(f"Meilleurs hyperparamètres pour {model_name}: {grid_search.best_params_}")
```

```python
from sklearn.metrics import roc_auc_score, classification_report, confusion_matrix, ConfusionMatrixDisplay

for model_name, model in best_models.items():
    y_pred = model.predict(X_test)
    print(f" Rapport de classification pour le modèle {model_name} :")
    print(classification_report(y_test, y_pred))
    ConfusionMatrixDisplay.from_predictions(y_test, y_pred)
    plt.title(f'Matrice de confusion - {model_name}')
    plt.show()
```

```python
# **Import des métriques**
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score, classification_report, confusion_matrix, ConfusionMatrixDisplay

# **Évaluation et comparaison des modèles**
model_performance = []

for model_name, model in best_models.items():
    print(f"Évaluation du modèle : {model_name}")
    
    # Prédictions sur le test set
    y_pred = model.predict(X_test)
    
    # Calcul des métriques
    accuracy = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    roc_auc = roc_auc_score(y_test, model.predict_proba(X_test)[:, 1])
    
    # Afficher le rapport de classification et la matrice de confusion
    print(f"Rapport de classification pour le modèle {model_name} :")
    print(classification_report(y_test, y_pred))
    
    ConfusionMatrixDisplay.from_predictions(y_test, y_pred)
    plt.title(f'Matrice de confusion - {model_name}')
    plt.show()
    
    # Enregistrer les performances dans la liste
    model_performance.append({
        'Modèle': model_name, 
        'Accuracy': accuracy, 
        'F1-Score': f1, 
        'ROC-AUC': roc_auc
    })

# Convertir la liste en DataFrame pour la comparaison
import pandas as pd
df_performance = pd.DataFrame(model_performance)

# Afficher la comparaison des performances des modèles
print("\n **Comparaison des performances des modèles**")
print(df_performance)
```

```python
# **Import des bibliothèques nécessaires**
from sklearn.metrics import RocCurveDisplay
import matplotlib.pyplot as plt

# **Tracer la courbe ROC pour chaque modèle**
for model_name, model in best_models.items():
    print(f"Courbe ROC pour le modèle : {model_name}")
    
    # Afficher la courbe ROC
    RocCurveDisplay.from_estimator(model, X_test, y_test)
    plt.title(f'Courbe ROC - {model_name}')
    plt.show()
```

```python
# **Import des bibliothèques nécessaires**
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier

# **Initialisation du modèle Random Forest**
rf_model = RandomForestClassifier(random_state=42, n_estimators=100, max_depth=None)

# **Validation croisée 5-Folds pour Random Forest**
cv_scores = cross_val_score(rf_model, X, y, cv=5, scoring='accuracy')

# **Affichage des résultats**
print(f'Scores de validation croisée (5-Folds) pour le modèle Random Forest: {cv_scores}')
print(f'Moyenne des scores: {np.mean(cv_scores)}')

# **Tracer les scores de validation croisée** 
plt.figure(figsize=(10, 6))
plt.plot(range(1, 6), cv_scores, marker='o', linestyle='--', color='b', label='Accuracy par Fold')
plt.axhline(np.mean(cv_scores), color='r', linestyle='-', label='Moyenne des Scores', linewidth=2)
plt.title('Scores de Validation Croisée 5-Folds pour Random Forest')
plt.xlabel('Fold')
plt.ylabel('Accuracy')
plt.xticks(range(1, 6))
plt.legend()
plt.show()
```

```python
# **Import des bibliothèques nécessaires**
import pandas as pd
import joblib  # ou pickle si vous avez utilisé pickle au lieu de joblib

# **1️- Charger le modèle Random Forest pré-entraîné**
loaded_model = joblib.load('random_forest_model.pkl')
print(' Modèle Random Forest chargé avec succès.')

# **2️- Charger le nouveau fichier CSV d'observations**
new_data = pd.read_csv(r"C:\Projet_ML\data\test.csv")
print(' Fichier chargé avec succès. Voici les 5 premières lignes :')
print(new_data.head())

#  **3️- Vérifier si la colonne "id" est présente et l'extraire**
if 'id' in new_data.columns:
    ids = new_data['id']
    new_data = new_data.drop(columns=['id'])
else:
    print(" La colonne 'id' n'a pas été trouvée dans le fichier d'entrée.")

#  **4️-Aligner les colonnes du fichier d'entrée avec celles du modèle**
# Récupérer les colonnes utilisées par le modèle lors de l'entraînement
model_columns = loaded_model.feature_names_in_

# 1️-Supprimer les colonnes inutiles (celles qui ne sont pas dans le modèle)
new_data = new_data[model_columns]

# 2️- Ajouter les colonnes manquantes qui étaient dans le modèle mais pas dans le fichier de prédiction
missing_cols = set(model_columns) - set(new_data.columns)
for col in missing_cols:
    new_data[col] = 0  # On peut aussi utiliser la médiane ou la moyenne au lieu de 0
print(f" Colonnes manquantes ajoutées avec des zéros : {missing_cols}")

#  **5️- Prédire les probabilités de fumer pour chaque individu**
probabilities = loaded_model.predict_proba(new_data)[:, 1]  # On récupère la probabilité de fumer (classe positive)

# **6️- Créer un DataFrame contenant les IDs et les probabilités**
results = pd.DataFrame({
    'id': ids,
    'smoking': probabilities
})

# **7️- Sauvegarder le fichier des résultats**
results.to_csv('predictions_fumeurs.csv', index=False)
print(' Fichier "predictions_fumeurs.csv" créé avec succès !')
print('Voici un aperçu des résultats :')
print(results.head())

```
